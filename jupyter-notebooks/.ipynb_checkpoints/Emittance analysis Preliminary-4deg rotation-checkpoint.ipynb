```python
# Define the Gaussian function
def gaussian(x, A, x0, sigma, c):
    return A * np.exp(-(x - x0)**2 / (2 * sigma**2)) + c
```

## Simulation Parameters

PP with a large hole 25 um, 200 um thickness, 4 deg rotation

**Sims parameters**

* source size = 2 um
* divergence 1.5 um
* energy 72 MeV

```python
file = uproot.open("../build-singlePP_4deg/Emittance_PP_experiment_params_72MeV_2um_1.85mrad.root")
# Emittance_PP_experiment_100MeV_5e6Events
# Emittance_PP_experiment_72MeV_5e6Events
# Emittance_PP_experiment_500MeV_5e6Events.root
```

```python
## Read data from Ntuple
key_id = 'Electrons_Screen;1'
ypos_tuple = file[key_id]['ypos'].array(library="np")
zpos_tuple = file[key_id]['zpos'].array(library="np")
xpos_tuple = file[key_id]['xpos'].array(library="np")

px_tuple = file[key_id]['px'].array(library="np")
py_tuple = file[key_id]['py'].array(library="np")
pz_tuple = file[key_id]['pz'].array(library="np")
#xpos = 180+1270 # in mm
```

```python
# hits at screen
scale_pos = 1 # mm
binsx = np.linspace(-scale_pos, scale_pos, 50)
binsy = np.linspace(-scale_pos, scale_pos, 50)
#
hist_position, yedges, zedges = np.histogram2d(ypos_tuple, zpos_tuple, bins=(binsx, binsy))
yedges_center = yedges[:-1] + (yedges[1] - yedges[2])/2
zedges_center = zedges[:-1] + (zedges[1] - zedges[2])/2

XX, YY = np.meshgrid(yedges_center, zedges_center)
```

```python
# divergence at screen
scale_theta = 0.7 # mm
binsx_theta = np.linspace(-scale_theta, scale_theta, 60)
binsy_theta = np.linspace(-scale_theta, scale_theta, 60)
#
hist_theta, yedges_theta, zedges_theta = np.histogram2d(ypos_tuple, zpos_tuple, bins=(binsx_theta, binsy_theta))
yedges_theta_center = yedges_theta[:-1] + (yedges_theta[1] - yedges_theta[2])/2
zedges_theta_center = zedges_theta[:-1] + (zedges_theta[1] - zedges_theta[2])/2

XX_theta, YY_theta = np.meshgrid(yedges_theta_center, zedges_theta_center)
```

```python
## vertical integration and Gaussian fit
# Integration
theta_vert_int = np.sum(hist_theta, axis = 0)

# Gaussian fit
initial_guess = [2000, 0, 0.2, 50]
popt_y, pcov_y = curve_fit(gaussian, yedges_theta_center, theta_vert_int, p0=initial_guess)
print(pcov_y)
print(popt_y)

data_fit_y = gaussian(yedges_theta_center, *popt_y)
```

```python
## vertical integration and Gaussian fit
# Integration
theta_horz_int = np.sum(hist_theta, axis = 1)

# Gaussian fit
initial_guess = [2000, 0, 0.2, 50]
popt_z, pcov_z = curve_fit(gaussian, zedges_theta_center, theta_horz_int, p0=initial_guess)
print(pcov_z)
print(popt_z)

data_fit_z = gaussian(zedges_theta_center, *popt_y)
```

```python
fig, axes1 = plt.subplots(2,2, figsize=(10, 8))
axes = iter(np.ravel(axes1))

###########################################
ax = next(axes)
ax.pcolormesh(XX, YY, hist_position, cmap='viridis')
ax.set_title('Hits at Screen')
ax.set_xlabel('y- mm')
ax.set_ylabel('z- mm')
ax.set_xlim(-0.7, 0.7)
ax.set_ylim(-0.7, 0.7)

###########################################
ax = next(axes)
ax.pcolormesh(XX_theta * 1e3, YY_theta*1e3, hist_theta, cmap='viridis')
ax.set_xlim(-600, 600)
ax.set_ylim(-600, 600)
ax.set_title('Divergence at the screen')
ax.set_xlabel('y- \u03BCrad')
ax.set_ylabel('z- \u03BCrad')

###########################################
ax = next(axes)
ax.plot(yedges_theta_center*1e3, theta_vert_int, label='GEANT4 Integration', color='black', lw=1.3)

ax.plot(yedges_theta_center*1e3, data_fit_y, label='Gaussian fit', color='red', lw=2 , linestyle='dashed')


text_legend2 = ("$\sigma_y$ = %.1f \u03BCrad\n" + 
                "$2\sigma_y$ = %.1f \u03BCrad\n"+
                "FWHM$_y$ = %.1f \u03BCrad\n")%(popt_y[2]*1e3, 
                                                2*popt_y[2]*1e3, 
                                                2.35*popt_y[2]*1e3)

_=ax.text(0.3, 0.35, text_legend2, horizontalalignment='left', color='black', fontsize=10, weight='normal',
          verticalalignment='center', transform=ax.transAxes)

ax.set_xlim(-600, 600)
#ax.set_ylim(-450, 450)
ax.set_title('$\\theta_y$ at the screen')
ax.set_xlabel('\u03BCrad')
ax.set_ylabel('Counts')

ax.legend(loc='lower right', fontsize=11, framealpha=0.95, facecolor='white')


###########################################
ax = next(axes)
ax.plot(zedges_theta_center*1e3, theta_horz_int, label='GEANT4 Integration', color='black', lw=1.3)

ax.plot(zedges_theta_center*1e3, data_fit_z, label='Gaussian fit', color='red', lw=2 , linestyle='dashed')


text_legendz = ("$\sigma_z$ = %.1f \u03BCrad\n" + 
                "$2\sigma_z$ = %.1f \u03BCrad\n"+
                "FWHM$_z$ = %.1f \u03BCrad\n")%(popt_z[2]*1e3, 
                                                2*popt_z[2]*1e3, 
                                                2.35*popt_z[2]*1e3)

_=ax.text(0.3, 0.35, text_legendz, horizontalalignment='left', color='black', fontsize=10, weight='normal',
          verticalalignment='center', transform=ax.transAxes)

ax.set_xlim(-600, 600)
#ax.set_ylim(-450, 450)
ax.set_title('$\\theta_z$ at the screen')
ax.set_xlabel('\u03BCrad')
ax.set_ylabel('Counts')

ax.legend(loc='lower right', fontsize=11, framealpha=0.95, facecolor='white')


plt.tight_layout()

fig.savefig('PP_Analysis_G4_4deg_4um_1.85mrad.png', dpi=450, facecolor='white', format='png', bbox_inches='tight')
plt.show()
```

```python
def fexp(x, a, b, c):
    return a * np.exp(x*b) + c

def fxx(x, a, b, c):
    return a * x**b + c
```

```python
popt_rel, pcov_rel  = curve_fit(fxx, source_size_um_sims, emittance_pi_mm_mrad, p0=[1, 5, 0])
print(pcov_rel)
print(popt_rel)
```

```python
xfit = np.linspace(0, 35, 100)
data_fit = fxx(xfit, *popt_rel)
```
