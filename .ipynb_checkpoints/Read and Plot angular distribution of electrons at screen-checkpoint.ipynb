## Load histograms

Example on how to open histograms

```python
## Load gamma phi histogram
bin_centers = file["h3;1"].axis().centers()
bin_values = file["h3;1"].values()
bin_errors = file["h3;1"].errors()
#

## plot
plt.step(bin_centers, bin_values, where='mid')
plt.yscale('log')
plt.xlabel('$\phi$')
plt.show()
```

```python
plt.polar(bin_centers*np.pi/180, bin_values)
plt.xlabel('$\phi$')
plt.ylim(0, 0.005)
plt.show()
```

## Load nTuples

Examples on how to get values from nTuples

```python
## Load ntuples 
print(file['Gamma;1'].typenames())
```

```python
## Read data from Ntuple
gamma_phi_tuple = file['Gamma;1']['phi'].array(library="np")
gamma_cosTheta_tuple = file['Gamma;1']['cos(theta)'].array(library="np")
```

```python
plt.hist2d(gamma_phi_tuple,gamma_cosTheta_tuple,  bins=(75,75), norm=LogNorm(1, 1e4))
plt.xlabel('$\phi$')
plt.ylabel('cos($\\theta$)')
plt.show()
```

```python
crit=(gamma_cosTheta_tuple>-0.2)*(gamma_cosTheta_tuple<0.2)
phi_f = gamma_phi_tuple[crit]
the_f = gamma_cosTheta_tuple[crit]

plt.hist2d(phi_f,the_f,  bins=(75,75), norm=LogNorm(1, 1e4))
plt.xlabel('$\phi$')
plt.ylabel('cos($\\theta$)')
plt.show()
```

```python
val1, bins1 = np.histogram(gamma_phi_tuple, bins=90)
val2, bins2 = np.histogram(phi_f, bins=90)

print ('%.2g' % len(gamma_phi_tuple) )
print ('%.2g' % len(phi_f) )
print ('%.2f %%' % (100*len(phi_f)/len(gamma_phi_tuple)) )


#plt.polar(bins1[:-1], val1)
plt.polar(bins2[:-1], val2)
#plt.ylim(0,50000)
plt.show()
```
